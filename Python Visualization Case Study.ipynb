```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SalesData = pd.read_csv(r'E:\Classes\Python\New folder\Python Foundation Case Study 4 - Python Visualization Case Study\Case Study 4 - Python Visualizations Case Study\SalesData.csv')
```

```python
SalesData.head()
```

```python
SalesData.info()
```

```python
region_sales = SalesData.groupby("Region")["Sales2015","Sales2016"].sum()
plt.figure(figsize=(12,6))
region_sales.plot(kind="bar",figsize=(8,8))
plt.ylabel('Sales')
plt.title("Sales by region for 2016 with 2015")
plt.show()
```

```python
sales_pie = SalesData.groupby("Region")["Sales2016"].sum()
plt.pie(sales_pie,autopct="%1.0f%%",labels=["Central","East","West"],shadow=True,explode=[0.0,0.1,0.0],colors=['r', 'g', 'b'])
plt.title("Contributing factors to the sales for each region in 2016")
plt.show()
```

```python
sales_region_tier = SalesData.groupby(["Region","Tier"])['Sales2015','Sales2016'].sum()
sales_region_tier.plot(kind="bar",figsize=(8,8))
plt.ylabel("Sales")
plt.title("Total sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

```python
sales_region_tier.plot(kind="barh",stacked=True,figsize=(8,8))
plt.xlabel("Sales")
plt.ylabel("Region/Tier")
plt.title("Total sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

```python
sales_state = SalesData.groupby(['Region',"State"])['Sales2015','Sales2016'].sum()
sales_east = sales_state.loc["East"]
sales_east.plot(kind="bar",figsize=(12,8),width=0.8)
plt.ylabel("Sales")
plt.title("Sales comparison between 2015 and 2016 for East Region")
plt.show()
```

```python
sales_division_tier =SalesData.groupby(["Tier","Division"])["Units2015","Units2016"].sum()
high_tier = sales_division_tier.loc["High"]
high_tier.plot(kind="bar",figsize=(12,8),width=0.7)
```

```python
month =SalesData["Month"]
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
SalesData["Qtr"]= pd.Series(quarter)
SalesData.head()
```

```python
qtr_sales = SalesData.groupby("Qtr")["Sales2015","Sales2016"].sum()
qtr_sales.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

```python
qtr_pivot = SalesData.pivot_table(index='Qtr',columns='Tier',values='Sales2016')
qtr_pivot
```

```python
plt.pie(x=qtr_pivot.loc["Q1",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
#Qtr Q2
plt.pie(x=qtr_pivot.loc["Q2",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```

```python
q3_data = qtr_pivot.loc["Q3", :]

# Make sure there are no negative values in the q3_data array
q3_data_non_negative = q3_data.abs()

plt.pie(q3_data_non_negative, autopct="%1.0f%%", labels=["High", "Low", "Med", "Out"], colors=['deepskyblue', 'darkorange', 'darkgreen', 'purple'])
plt.title("Tier Distribution in Q3")
plt.show()
```

```python
#Qtr Q4
plt.pie(x=qtr_pivot.loc["Q4",:],autopct="%1.0f%%",labels=["High","Low","Med","Out"],colors=['deepskyblue', 'darkorange', 'darkgreen'])
plt.show()
```
